```python
def Yangon ():
    Combined = list (zip(Citylist,Productlinelist,totallist))
    YangonListj = []
    YangonListk = []
    for i,j,k in Combined:
        if i == "Yangon":
            YangonListj.append (j)
            YangonListk.append (k)
    YangonList = list(zip(YangonListj,YangonListk))
    HealthandBeauty = []
    Electronicaccessories = []
    Homeandlifestyle = []
    Sportsandtravel = []
    Foodandbeverages = []
    Fashionaccessories = []
    for h,l in YangonList:
        if h == "Health and beauty":
            HealthandBeauty.append (l)
        if h == "Electronic accessories":
            Electronicaccessories.append (l)
        if h == "Home and lifestyle":
            Homeandlifestyle.append (l)
        if h == "Sports and travel":
            Sportsandtravel.append (l)
        if h == "Food and beverages":
            Foodandbeverages.append (l)
        if h == "Fashion accessories":
            Fashionaccessories.append (l)

    SumHealthandBeauty = sum(HealthandBeauty)
    SumElectronicaccessories = sum(Electronicaccessories)
    SumHomeandlifestyle = sum(Homeandlifestyle)
    SumSportsandtravel = sum(Sportsandtravel)
    SumFoodandbeverages = sum(Foodandbeverages)
    SumFashionaccessories = sum(Fashionaccessories)

    print ("Health and beauty in Yangon:",SumHealthandBeauty)
    print ("Electronic accessories in Yangon:",SumElectronicaccessories)      
    print ("Home and lifestyle in Yangon:",SumHomeandlifestyle) 
    print ("Sports and travel in Yangon:",SumSportsandtravel)
    print ("Food and beverages in Yangon:",SumFoodandbeverages)
    print ("Fashion accessories in Yangon:",SumFashionaccessories)

    MostSoldYangon = max(SumHealthandBeauty,SumElectronicaccessories,SumHomeandlifestyle,SumSportsandtravel,SumFoodandbeverages,SumFashionaccessories)
    print("The most sold product line in Yangon:", MostSoldYangon)
Yangon ()
```

```python
time = mydata ["Time"]
timelist = list (time)

quantity = mydata ["Quantity"]
quantitylist = list (quantity)

timelist = [time.replace (":",".") for time in timelist]


timelist = [float (time) for time in timelist]


combined = list (zip(timelist,quantitylist))

FirstQuadrant = []
SecondQuadrant = []
ThirdQuadrant = []
FourthQuadrant = []

for time,quantity in combined:
    if time >= 0 and time < 6:
        FirstQuadrant.append (quantity)
    if time >= 6 and time < 12:
        SecondQuadrant.append (quantity)
    if time >= 12 and time < 18:
        ThirdQuadrant.append (quantity)
    if time >= 18 and time < 24:
        FourthQuadrant.append (quantity)
    

print (sum (FirstQuadrant))
print (sum (SecondQuadrant))
print (sum (ThirdQuadrant))
print (sum (FourthQuadrant))   
       
        
```
